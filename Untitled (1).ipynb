```python
#Bar Chart for Categorical Features
#Pclass
#Sex
#SibSp ( # of siblings and spouse)
#Parch ( # of parents and children)
#Embarked
#Cabin
```

```python
def bar_chart(feature):
    survived = train_data[train_data['Survived']==1][feature].value_counts()
    dead = train_data[train_data['Survived']==0][feature].value_counts()
    df = pd.DataFrame([survived,dead])
    df.index = ['Survived','Dead']
    df.plot(kind='bar',stacked=True, figsize=(10,5))
```

```python
bar_chart('Pclass')
```

The Chart confirms 1st class more likely survivied than other classes
The Chart confirms 3rd class more likely dead than other classes

```python
bar_chart('Sex')
```

female more likely than male 

```python
bar_chart('SibSp')
```

The Chart confirms a person aboarded with more than 2 siblings or spouse more likely survived
The Chart confirms a person aboarded without siblings or spouse more likely dead

```python
bar_chart('Parch')
```

The Chart confirms a person aboarded with more than 2 parents or children more likely survived
The Chart confirms a person aboarded alone more likely dead

```python
bar_chart('Embarked')
```

The Chart confirms a person aboarded from C slightly more likely survived

The Chart confirms a person aboarded from Q more likely dead

The Chart confirms a person aboarded from S more likely dead

```python
#combining train and test dataset

all_data=[test_data,train_data]
```

```python
#title map
for dataset in all_data:
    dataset['Title']=dataset['Name'].str.extract(' ([A-Za-z]+)\.', expand=True)
```

```python
train_data['Title'].value_counts()
```

```python
#mapping for converting name to Title
title_mapping = {"Mr": 0, "Miss": 1, "Mrs": 2, 
                 "Master": 3, "Dr": 3, "Rev": 3, "Col": 3, "Major": 3, "Mlle": 3,"Countess": 3,
                 "Ms": 3, "Lady": 3, "Jonkheer": 3, "Don": 3, "Dona" : 3, "Mme": 3,"Capt": 3,"Sir": 3 }
for dataset in all_data:
    dataset['Title'] = dataset['Title'].map(title_mapping)      
```

```python
train_data.head()
```

```python
#drop name
train_data.drop('Name', axis=1, inplace=True)
test_data.drop('Name', axis=1, inplace=True)
```

```python
train_data.head()
```

```python
#sex mapping
Sex_mapping={"male":0,"female":1}
for i in all_data:
   i['Sex'] = i['Sex'].map(Sex_mapping)
    
```

```python
train_data.head()
```

```python
#missing age
#use Title's median age for missing Age
train_data["Age"].fillna(train_data.groupby("Title")["Age"].transform("median"), inplace=True)
test_data["Age"].fillna(test_data.groupby("Title")["Age"].transform("median"), inplace=True)
```

```python
train_data.head(100)
train_data.groupby("Title")["Age"].transform("median")
```

```python
#Converting Numerical Age to Categorical Variable
train_data['AgeBand'] = pd.cut(train_data['Age'], 5)
train_data[['AgeBand', 'Survived']].groupby(['AgeBand'], as_index=False).mean().sort_values(by='AgeBand', ascending=True)
```

```python
for dataset in all_data:   
    dataset.loc[ dataset['Age'] <= 16, 'Age'] = 0
    dataset.loc[(dataset['Age'] > 16) & (dataset['Age'] <= 32), 'Age'] = 1
    dataset.loc[(dataset['Age'] > 32) & (dataset['Age'] <= 48), 'Age'] = 2
    dataset.loc[(dataset['Age'] > 48) & (dataset['Age'] <= 64), 'Age'] = 3
    dataset.loc[ dataset['Age'] > 64, 'Age']
train_data.head(10)
```

```python
test_data.head()
```

```python
#drop AgeBand
train_data.drop('AgeBand',axis=1,inplace=True)
train_data.head()
```

```python
#combining all family data in FamilySize
train_data["FamilySize"] = train_data["SibSp"] + train_data["Parch"] + 1
test_data["FamilySize"] = test_data["SibSp"] + test_data["Parch"] + 1

train_data[['FamilySize', 'Survived']].groupby(['FamilySize'], as_index=False).mean().sort_values(by='Survived', ascending=False)
```

```python
test_data.head(10)
```

```python
#drop unnecessary data 
train_data = train_data.drop(['SibSp','Parch' ], axis=1)
test_data = test_data.drop(['SibSp','Parch'], axis=1)
#all_data = [train_data, test_data]

train_data.head()
```

```python
test_data.head(20)
```

```python
#missing data in test 
test_data['Fare'].fillna(test_data['Fare'].median(), inplace=True)
test_data.head()
```

```python
#Converting Numerical Age to Categorical Variable
train_data['FareBand'] = pd.qcut(train_data['Fare'], 4)
train_data[['FareBand', 'Survived']].groupby(['FareBand'], as_index=False).mean().sort_values(by='FareBand', ascending=True)
```

```python
#Convert the Fare feature to ordinal values based on the FareBand
for dataset in all_data:
    dataset.loc[ dataset['Fare'] <= 7.91, 'Fare'] = 0
    dataset.loc[(dataset['Fare'] > 7.91) & (dataset['Fare'] <= 14.454), 'Fare'] = 1
    dataset.loc[(dataset['Fare'] > 14.454) & (dataset['Fare'] <= 31), 'Fare']   = 2
    dataset.loc[ dataset['Fare'] > 31, 'Fare'] = 3
    dataset['Fare'] = dataset['Fare'].astype(int)
```
